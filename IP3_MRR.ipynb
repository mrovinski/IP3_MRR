**Problem 1:** Write a function that takes a filename and returns a dataframe  The function should convert the Date column to a date and add a column called company.

```python
def open_dataframes(file):
    data = pd.read_csv(file)
    df = pd.DataFrame(data)
    df.Date = pd.to_datetime(df.Date)
    df.insert(0, 'Company', file[:-4])
    return df
```

**Problem 6:** Combine the two bar graphs into a single graph that shares the x-axis. It will be published in the newspaper in grayscale so the top graph should be silver and the bottom should be gray. Save the file as 'combined_bar_chart.png'.

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, sharex=True)
fig.subplots_adjust(wspace=0.3, hspace=0.5)
ax2 = fig.add_subplot(212,  sharex=ax1, sharey=ax2)
Noon.sort_values().plot.barh(color = 'gray', xlabel='Company and Year', title='TW/hours by Company/Year @ Noon')
ax1= fig.add_subplot(211, sharex=ax2, sharey=ax1)
Midnight.sort_values().plot.barh(color = 'silver', xlabel='Company and Year', title='TW/hours by Company/Year @ Midnight')
plt.savefig('combine_bar_chart.png')
```

**Problem 7:** Write a function that accepts the ungrouped dataframe, the company name, date, and a plotting color that generates an area graph for that particular company/day.

```python
def daily_views(dataframe, company, date, color):
    q = "Company=='"+company+"' and Date=='"+date+"'" 
    return dataframe.query(q).iloc[:,2:].T.plot(kind = 'area', 
                                                xlabel= 'Time of Day', 
                                                ylabel = 'TW/h Generated', 
                                                title = 'Power Production:'+company+ '/' +date,
                                                color = color, 
                                                legend = None)
```

```python
daily_views(df, '50Hertz', '2019-08-24', 'purple')
```

**Bonus:** Generate a unified graph that contains four area charts, on for each company for the same day. 

```python
fig = plt.figure(figsize = (12,7))
fig.subplots_adjust(wspace = 0.3, hspace = 0.4)
ax1 = fig.add_subplot(221)
df.query("Company == '50Hertz' and Date == '2019-08-24'").iloc[:,2:].T.plot(kind='area', ax=ax1, color = 'red', xlabel = 'Time of Day', ylabel = 'TW/h Generated', title = 'Power Production: 50Hertz/2019-08-24', legend = None)
ax2 = fig.add_subplot(222)
df.query("Company == 'Amprion' and Date == '2019-08-24'").iloc[:,2:].T.plot(kind='area', ax=ax2, color = 'green', xlabel = 'Time of Day', ylabel = 'TW/h Generated', title = 'Power Production: Amprion/2019-08-24', legend = None)
ax3 = fig.add_subplot(223)
df.query("Company == 'TenneTTSO' and Date == '2019-08-24'").iloc[:,2:].T.plot(kind='area', ax = ax3, color = 'blue', xlabel = 'Time of Day', ylabel = 'TW/h Generated', title = 'Power Production: TenneTTSO/2019-08-24', legend = None)
ax4 = fig.add_subplot(224)
df.query("Company == 'TransnetBW' and Date == '2019-08-24'").iloc[:,2:].T.plot(kind='area', ax = ax4, color = 'purple', xlabel = 'Time of Day', ylabel = 'TW/h Generated', title = 'Power Production: TransnetBW/2019-08-24', legend = None)
```
